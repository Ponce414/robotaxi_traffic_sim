# Robo-Taxi Adoption Dynamics

This notebook explores how the share of robo-taxis (RT) evolves over time in the LA traffic simulation model. We will use the shared core API to run scenarios with varying robo-taxi shares and visualize the impact on key metrics such as total VMT, average trip time, and congestion.

---

**Outline:**
1. Import required libraries and model API
2. Run scenarios for different robo-taxi shares
3. Plot and analyze results


```python
%pip install matplotlib numpy

from pathlib import Path
import sys
import matplotlib.pyplot as plt
import numpy as np

# Ensure the project root (which contains graph_model.py) is in sys.path
PROJECT_ROOT = Path.cwd().resolve()
while not (PROJECT_ROOT / "graph_model.py").exists() and PROJECT_ROOT != PROJECT_ROOT.parent:
	PROJECT_ROOT = PROJECT_ROOT.parent

if not (PROJECT_ROOT / "graph_model.py").exists():
	raise FileNotFoundError("Could not find 'graph_model.py' in any parent directory.")

if str(PROJECT_ROOT) not in sys.path:
	sys.path.insert(0, str(PROJECT_ROOT))

# Import required libraries and model API
from graph_model import SimParams, build_graph, run_scenario, get_baseline_vmt

# Build the LA road network graph
G = build_graph()
```

## Simulate Scenarios with Varying Robo-Taxi Share

We will run the model for a range of `robotaxi_share` values from 0 to 1 and record key metrics for each scenario.

```python
%pip install scipy

# Define range of robo-taxi shares
rt_shares = np.linspace(0, 1, 11)
results = []

for share in rt_shares:
    params = SimParams(robotaxi_share=share)
    result = run_scenario(G, params)
    results.append({
        'robotaxi_share': share,
        'total_vmt': result['total_vmt'],
        'avg_trip_time_min': result['avg_trip_time_min'],
        'congested_edge_count': result['congested_edge_count'],
        'unmet_demand': result['unmet_demand'],
    })
```

## Plot Results

We will visualize how total VMT, average trip time, and congestion change as robo-taxi share increases.

```python
# Convert results to arrays for plotting
robotaxi_shares = [r['robotaxi_share'] for r in results]
total_vmt = [r['total_vmt'] for r in results]
avg_trip_time = [r['avg_trip_time_min'] for r in results]
congested_edges = [r['congested_edge_count'] for r in results]
unmet_demand = [r['unmet_demand'] for r in results]

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].plot(robotaxi_shares, total_vmt, marker='o')
axs[0, 0].set_title('Total VMT vs Robo-Taxi Share')
axs[0, 0].set_xlabel('Robo-Taxi Share')
axs[0, 0].set_ylabel('Total VMT')

axs[0, 1].plot(robotaxi_shares, avg_trip_time, marker='o', color='orange')
axs[0, 1].set_title('Avg Trip Time vs Robo-Taxi Share')
axs[0, 1].set_xlabel('Robo-Taxi Share')
axs[0, 1].set_ylabel('Avg Trip Time (min)')

axs[1, 0].plot(robotaxi_shares, congested_edges, marker='o', color='green')
axs[1, 0].set_title('Congested Edges vs Robo-Taxi Share')
axs[1, 0].set_xlabel('Robo-Taxi Share')
axs[1, 0].set_ylabel('Congested Edge Count')

axs[1, 1].plot(robotaxi_shares, unmet_demand, marker='o', color='red')
axs[1, 1].set_title('Unmet Demand vs Robo-Taxi Share')
axs[1, 1].set_xlabel('Robo-Taxi Share')
axs[1, 1].set_ylabel('Unmet Demand')

plt.tight_layout()
plt.show()
```

## Discussion

- **Total VMT**: Shows how vehicle miles traveled changes as robo-taxi share increases.
- **Average Trip Time**: Indicates the impact on user travel time.
- **Congested Edges**: Reflects network congestion.
- **Unmet Demand**: Shows if the fleet cap is binding at high robo-taxi shares.
